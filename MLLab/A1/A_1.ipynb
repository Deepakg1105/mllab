**Mean, Median, Mode, Std. Deviation, Variance, min-max normalization, standardization ?**



```python
data = [115.3,195.5,120.5,110.2,90.4,105.6,110.9,116.3,122.3,125.4]
```

```python
sum = 0
for i in data:
  sum+=i

mean = sum/len(data)
```

```python
print(mean)
```

```python
data.sort()
data
```

```python
median = 0
if len(data)%2==0:
  median = 0.5 * (data[len(data)//2] + data[len(data)//2 - 1])
else :
  median = data[len(data)//2]
```

```python
median
```

```python
max_count = 0
dict = {}
for i in data:
  dict[i] = data.count(i)
  
```

```python
for element, count in dict.items():
  if count>max_count:
    max_count = count
  
print("Mode(s) is/are")
for element, count in dict.items():
  if count==max_count:
    print(element)

```

**Std Deviation**

image.png

```python
sum=0
for i in data:
  sum = sum + (i-mean)**2
variance = sum/(len(data)-1)
variance
```

```python
std_dev = variance**0.5
std_dev
```

**Min-max Normalization**

image.png

```python
min_max_normal = []
min=data[0]
max=0
for i in data:
  if min>i:
    min=i

  if max<i:
    max=i

for i in data:
  x_dash = (i-min)/(max-min)
  min_max_normal.append(x_dash)

min_max_normal

```

**Standardization**

image.png

```python
standardized_data = []

for i in data:
  x_std = (i-mean)/std_dev
  standardized_data.append(x_std)

standardized_data
```
